### handling missing values:

```python
# Fill numeric columns with median
for col in ['income_annum', 'loan_amount', 'cibil_score']:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
for col in ['education', 'self_employed']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.columns = df.columns.str.strip()

# Show missing values before handling
print("Missing values BEFORE handling:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Handle missing values
# Fill numeric columns with median
numeric_cols = ['income_annum', 'loan_amount', 'cibil_score']
for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
categorical_cols = ['education', 'self_employed']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

# Show missing values after handling
print("\nMissing values AFTER handling:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

## handeling Outliers: 

```python
numeric_cols = ['income_annum', 'loan_amount', 'cibil_score']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
df.columns = df.columns.str.strip()

numeric_cols = ['income_annum', 'loan_amount', 'cibil_score']

print("Outliers detection and handling with IQR method:\n")

for col in numeric_cols:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers count
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
        print(f"{col} - Outliers count: {outliers_count}")

        print(f"Before handling {col} summary:")
        print(df[col].describe())

        # Handle outliers by capping
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

        print(f"After handling {col} summary:")
        print(df[col].describe())
        print("-" * 40)
```
